### Changeing Image_urls path

```python
def Image_url(file_name,file_path):
    
    new = []
    
    #Reading Excel file
    df = pd.read_excel(file_name)
    
    #Reading Csv file
    #df = pd.read_csv(file_name)

    for urls in df["Image_Url"]:
        image_name = urls.split("\\")[-1]
        new.append(os.path.join(file_path,image_name))

    #Replaceing values in the Image_url into new Urls
    df["Image_Url"] = new
    
    return df
```
